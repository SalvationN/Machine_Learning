```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test = pd.read_csv('data/RentListingInquries_FE_test.csv')
train = pd.read_csv('data/RentListingInquries_FE_train.csv')
train.head()
```

```python
y_train = train['interest_level']
x_train = train.drop('interest_level',axis=1)
```

数据已经经过特征工程处理，所以接下来只需要掉包......(开心)

# 缺省参数的决策树

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score
from sklearn.metrics import log_loss
dtree = DecisionTreeClassifier()
dtree.fit(x_train,y_train)
y_pred = dtree.predict_proba(x_train)
print('训练集上的正确率为：',dtree.score(x_train,y_train))
print('训练集上的log_loss为：',log_loss(y_train,y_pred))
```

```python
dtree.get_params
```

```python
df = pd.DataFrame({'columns':x_train.columns,'importance':dtree.feature_importances_.T})
df.sort_values(by=['importance'],ascending=False)
```

按照重要性对特征进行排序，取大于0的特征看看

```python
df = df[df.importance>0]
df
```

想要选取一些重要的特征再进行建模的话，可能要把特征数目调少一些

```python
df = df[df.importance>0.005]
df
```

```python
from numpy import sort
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
# Fit model using each importance as a threshold
thresholds = sort(df.importance)
for thresh in thresholds:
  # select features using threshold
  selection = SelectFromModel(dtree, threshold=thresh, prefit=True)
  select_x_train = selection.transform(x_train)

  # train model
  selection_model = DecisionTreeClassifier()
  selection_model.fit(select_x_train, y_train)
    
# eval model
  select_x_train = selection.transform(x_train)
  y_pred = selection_model.predict(select_x_train)
  predictions = [round(value) for value in y_pred]
  accuracy = accuracy_score(y_train, predictions)
  print("Thresh=%.3f, n=%d, Accuracy: %.2f%%" % (thresh, select_x_train.shape[1],
      accuracy*100.0))
```

经过特征选择，发现当thresh为0.07时的结果都还是挺好的

# 决策树参数调优

```python
from sklearn.model_selection import GridSearchCV
max_depth = range(1,20)
min_samples_leaf = range(1,30)
dtree2 = DecisionTreeClassifier()
param_grid = {'max_depth':max_depth,'min_samples_leaf':min_samples_leaf}
grid = GridSearchCV(dtree2,param_grid=param_grid,cv=5,scoring='accuracy')
grid.fit(x_train,y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```
