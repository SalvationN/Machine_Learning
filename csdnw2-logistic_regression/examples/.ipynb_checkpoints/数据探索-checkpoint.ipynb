```python
#导入必要工具包
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

这些特征的方差都有点大，经观察，可能是因为零值过多。有些特征有零值是正常的，但如胰岛素含量、身高体重比这些特征是不可能有零值的，所以，需要对这些特征做处理

# 查看每个变量的分布

## Outcome

```python
sns.countplot(data.Outcome)
plt.xlabel('Diabetes')
plt.ylabel('Outcome')
```

未得病的样本一些，符合逻辑

## Pregnancies

```python
sns.countplot(data.Pregnancies)
plt.xlabel('number of pregnancies')
plt.ylabel('number of occurancies')
```

17次怀孕.....有点夸张的，但是考虑到怀孕次数越多，得病的概率是不是会越高呢？

```python
sns.countplot(x='Pregnancies',hue='Outcome',data=data)
```

果然，和猜测的一样，怀孕次数越多得病的概率越高

## Glucose

```python
sns.distplot(data.Glucose)
```

```python
sns.violinplot(x='Outcome',y='Glucose',hue='Outcome',data=data)
```

显然，得病与不得病时，葡萄糖浓度的分布是不同的，总体来说偏高很多

## BloodPressure

```python
sns.distplot(data.BloodPressure)
```

有明显突兀的零值存在，因此，这个特征里的零值是必须要消灭的

```python
sns.violinplot(x='Outcome',y='BloodPressure',hue='Outcome',data=data)
```

直观上看，血压在得病与不得病时的分布相差无几

## SkinThickness	

```python
sns.distplot(data.SkinThickness	)
```

这个特征也一样，零值过多，而且多的有点过分

```python
sns.violinplot(x='Outcome',y='SkinThickness',hue='Outcome',data=data)
```

抛开零值不谈，总的来说，得病时该特征值稍微高一些

```python
sns.distplot(data.Insulin)
```
